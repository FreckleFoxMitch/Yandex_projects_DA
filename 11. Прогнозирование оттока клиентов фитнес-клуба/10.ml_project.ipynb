# Прогнозирование оттока клиентов фитнес-клуба

### Описание проекта
Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию взаимодействия с клиентами на основе аналитических данных. Распространённая проблема фитнес-клубов и других сервисов — отток клиентов. Для фитнес-центра можно считать, что клиент попал в отток, если за последний месяц ни разу не посетил спортзал. Наша задача — провести анализ и подготовить план действий по удержанию клиентов.

**Цели исследования:**
1. *Научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента.*
2. *Сформировать типичные портреты клиентов.*
3. *Проанализировать признаки, наиболее сильно влияющие на отток.*

**Описание данных**

*Данные клиента за предыдущий до проверки факта оттока месяц:*

    gender — пол;
    Near_Location — проживание или работа в районе, где находится фитнес-центр;
    Partner — сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);
    Promo_friends — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента);
    Phone — наличие контактного телефона;
    Age — возраст;
    Lifetime — время с момента первого обращения в фитнес-центр (в месяцах).
    
*Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:*

    Contract_period — длительность текущего действующего абонемента (месяц, 6 месяцев, год);
    Month_to_end_contract — срок до окончания текущего действующего абонемента (в месяцах);
    Group_visits — факт посещения групповых занятий;
    Avg_class_frequency_total — средняя частота посещений в неделю за все время с начала действия абонемента;
    Avg_class_frequency_current_month — средняя частота посещений в неделю за предыдущий месяц;
    Avg_additional_charges_total — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон.

    Churn — факт оттока в текущем месяце.

### Гистограммы распределения признаков в группах по факту оттока
Определим общие для всех графиков параметры визуализации:

```python
sns.set(
    font_scale = 1.4,
    style = 'darkgrid'
)
```

Для каждого признака построим графики распределения с плотностью распределения в группах по факту оттока:

## Модель прогнозирования оттока клиентов

Эта задача относится к *обучению с учителем* и может быть реализована несколькими алгоритмами. Модели, обученные ими, будут не просто прогнозировать финальное значение факта оттока для клиента, а ещё и оценивать вероятность рассматриваемого события. В данном исследовании мы будем использовать алгоритмы логистической регрессии `LogisticRegression()` и случайного леса `RandomForestClassifier()` из библиотеки `sklearn`.

### Разделение данных
Сперва разделим наши данные из `gym` на признаки (матрица `X`) и целевую переменную (`y`):

```python
X = gym.drop(['churn'], axis=1)
y = gym.churn
```

Целевая переменная `churn` не имеет временной структуры и задачи прогнозирования временного ряда перед нами не стоит. Поэтому подойдёт случайное разбиение выборки.

Разделим данные на обучающую и валидационную выборки подходящим для этого методом `train_test_split()` в пропорции 80:20. Для воспроизводимости результата укажем аргумент `random_state=0`:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Код ревьюера

print(X_train.shape, X_test.shape)
```

### Стандартизация данных
На этапе EDA мы заметили, что все признаки представлены в разных масштабах. Многие алгоритмы машинного обучения лучше работают на стандартизированных данных, где значения признаков соответствуют нормальному распределению. 

Стандартизируем данные методом `StandardScaler()`. Для начала создадим объект этого класса:

```python
scaler = StandardScaler()
```

### Обучение моделей
Создадим список моделей `models` для упомянутых выше алгоритмов.

Поскольку в будущем в `LogisticRegression()` алгоритм решения по умолчанию (солвер) будет изменён, мы укажем аргумент `solver='liblinear'`.

В `RandomForestClassifier()` все параметры дерева (в том числе и их число `n_estimators`) оставим по умолчанию. Посмотрим, как это повлияет на итоговый результат.

В обоих случаях для воспроизводимости результата укажем аргумент `random_state=0`:

```python
models = [
    LogisticRegression(solver='liblinear', random_state=0),
    RandomForestClassifier(random_state=0)
]
```

В задачах бинарной классификации оцениваются следующие метрики:
* доля правильных ответов `accuracy_score`
* точность `precision_score`
* полнота `recall_score`

Однако тут стоит оговорить несколько моментов:


1. Доля правильных ответов `accuracy_score` хорошо работает только в условиях баланса классов 50:50. В рассматриваемом случае баланс смещён в сторону `0` (т.е. остаются клиенты гораздо чаще, чем уходят в отток). Поэтому оценивать модель исключительно по данной метрике не стоит.


2. Точность `precision_score` и полнота `recall_score` к соотношению классов не привязаны. Первая говорит, какая доля прогнозов относительно `1` верна, а вторая показывает, сколько реальных объектов `1` модель смогла обнаружить. То есть, характеристики разнонаправлены. Поскольку в нашем случае мы не хотим, чтобы модель срабатывала ложно (причислять к оттоку потенциально лояльных и платёжеспособных клиентов чревато потерей прибыли), то больше внимания стоит сосредоточить именно на `precision_score`.


3. Ввиду того, что `precision_score` и `recall_score` направлены на избежание противоположных рисков, нужна сводная метрика, учитывающая баланс между метриками - `f1_score`. Поэтому модели стоит сравнивать в том числе и по ней.


4. Площадь под кривой ошибок `roc_auc_score` стоит немного в стороне: она показывает не столько ответ для каждого наблюдения (`0` или `1`), как это делают метрики выше, сколько вероятность отнесения к тому или иному классу. Она отвечает на вопрос, насколько хорошо будут выделяться реальные классы объектов, хорошо работает на несбалансированных классах - и поэтому хорошо подходит для оценки качества моделей.


Напишем функцию `make_prediction`, которая принимает на вход модель и данные и выводит эти метрики:

```python
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    print(
        'Accuracy: {:.2f}\nPrecision: {:.2f}\nRecall: {:.2f}\nF1: {:.2f}\nROC-AUC: {:.2f}'.format(
            accuracy_score(y_test, y_pred),
            precision_score(y_test, y_pred),
            recall_score(y_test, y_pred),
            f1_score(y_test, y_pred),
            roc_auc_score(y_test, y_proba)
            
        )
    )
```

В цикле выведем метрики моделей:

```python
for m in models:
    print(m)
    make_prediction(m, X_train_st, y_train, X_test_st, y_test)
    print('--------')
```

Исходя из близким к единице значениям метрик, можно заключить, что обе модели работают хорошо и переобучения ни у одной из них не наблюдается. 

По метрике точности `precision_score` и сводной метрике `f1_score` логистическая регрессия показывает себя чуть лучше случайного леса. По метрике площади под кривой ошибок `roc_auc_score` обе модели работают одинаково хорошо (т.е. хорошо выделяют разные классы).

В целом результат нас устраивает. Однако при необходимости можно поэкспериментировать с классификатором случайного леса, добавив в модель `n_estimators = 100`, - и посмотреть, изменятся ли в лучшую сторону интересующие нас метрики.

## Кластеризация клиентов
Модели прогнозирования оттока в предыдущем пункте проявили себя очень хорошо. Однако в реалиях бизнеса нам важно не только верно предсказывать факт оттока, но и понять, на какие условные группы можно разделить клиентов, чтобы сделать политику фитнес-центра более персонализированной - и тем самым этот отток минимизировать. Т.е. перед нами встаёт уже задача кластеризации.

Кластеризация относится к *обучению без учителя* и может быть реализована алгоритмами:
* K-Means
* Агломеративной иерархической кластеризацией

Поскольку K-Means чувствителен к заданному числу кластеров, а агломеративная иерархическая кластеризация как раз позволяет их визуализировать, то начать стоит именно с неё. Однако сперва признаки необходимо привести к единому масштабу методом стандартизации.

### Стандартизация данных
Стандартизируем матрицу признаков методом `StandardScaler()`. Для начала создадим объект этого класса:

```python
sc = StandardScaler()
```

Применим его к матрице признаков `X`. Обучим `sc` и одновременно трансформируем её:

```python
X_sc = sc.fit_transform(X)
```

### Построение матрицы расстояний и дендрограммы
Матрица расстояний и дендрограмма будут строиться при помощи алгоритма агломеративной иерархической кластеризации. 

Ранее из библиотеки `scipy.cluster.hierarchy` были импортированы классы моделей `linkage()` и `dendrogram()` - для матрицы расстояний и дендрограммы, соответственно. Передадим стандартизированную таблицу `X_sc` в качестве параметра функции `linkage()`. Чтобы
минимизировать варианты расстояний между кластерами, передадим параметру `method`
значение `'ward'`:

```python
linked = linkage(X_sc, method = 'ward')
```

В переменной `linked` сохранена таблица «связок» между объектами. Визуализируем её как дендрограмму:

```python
plt.figure(figsize=(15, 10))
dendrogram(linked, orientation='top')
plt.title('Иерархическая кластеризация')
plt.show()
```

Как мы видим, модель предложила нам сегментацию клиентов на 4 кластера. При этом четвёртый кажется слишком большим. Предпочтительней всего будет провести границу на условной отметке в 60 - и таким образом **выделить 5 кластеров**.

Расчёты попарных расстояний занимают очень много времени. Поэтому мы используем определённое выше значение числа кластеров - и запустим более быстрый алгоритм K-Means.

### Кластеризация методом K-Means
Алгоритм под названием `KMeans` реализован в модуле `sklearn.cluster`, его мы импортировали заранее.

Зададим число кластеров, равное 5 (выше мы уже указывали, почему), и зафиксируем значение параметра `random_state=0`:

```python
km = KMeans(n_clusters=5, random_state=0)
```

Применим алгоритм к стандартизированным данным и сформируем вектор кластеров `labels`:

```python
labels = km.fit_predict(X_sc)
```

В `labels` сохранены индексы предложенных алгоритмом групп. Алгоритм случайно назначает номер определённой группе, поэтому искать в этой цифре какой-то смысл не нужно: группа с индексом 2 не «ближе» к группе с индексом 3, чем группа 1.

Добавим `labels` как столбец к датафрейму `gym`:

```python
gym['cluster_km'] = labels
```

  По данному количеству кластеров была построена кластеризация методом K-Means. Метки кластеров были добавлены к основному датафрейму.
  
  
* *Cредние значения признаков для кластеров*

  Исходя из средних значений, было выявлено, что:

  * В кластер `0` попали преимущественно живущие поблизости клиенты, б*о*льшая часть которых являются сотрудниками компаний-партнёров клуба. Длительность их абонементов - порядка года, они много (в среднем 160.8) тратят на дополнительные услуги клуба. Среднее "время жизни" составляет порядка 4.7 месяца, а клуб они посещают примерно дважды в неделю.

  * Клиенты кластера `1` отличаются полным отсутствием контактного телефона, только половина из них - участники партнёрской программы. Длительность их абонемента в среднем - около полугода, и около 5 месяцев до его окончания (т.е. это клиенты, которые ходили на занятия всего месяц в общей сумме). Их средние траты на дополнительные услуги - 143.9, среднее "время жизни" - около 3.9 месяцев. В предыдущем месяце в среднем они посещали клуб реже, чем за весь период.

  * Весь кластер `2` - это клиенты, живущие/работающие далеко от клуба. Возможно по этой причине они почти не участвуют в промо-акции (всё зависит от условий её проведения). Они оформляют абонементы чаще на месяц (хотя и полугодовые тоже встречаются) и редко посещают групповые занятия. До окончания действия абонемента у них в среднем около 2 месяцев, "время жизни" - около 3 месяцев.

  * Абсолютно все клиенты из кластера `3` - местные жители либо работающие рядом. Чаще всего они оформляют абонементы на месяц (реже - на полгода). До окончания действия абонемента у них в меньше 2 месяцев, "время жизни" - 2.5 месяца. Они меньше всех прочих тратят на дополнительные услуги (131.6) и в прошедшем месяце реже всех появлялись на занятиях (около раза в неделю).

  * И наконец, кластер `4` - самый щедрый с точки зрения трат на дополнительные услуги (в среднем 161.6) и самый взрослый (средний возраст - 30 лет). Б*о*льшая часть живут или работают неподалёку, оформляют абонементы чаще на месяц или полгода, а их "время жизни" даже выше, чем у самого лояльного кластера `0` - 4.9 месяца. Также они чаще прочих посещают центр - около 3 раз в неделю, как за всё время, так и за предыдущий месяц.
  
  
* *Гистограммы распределения признаков для кластеров*

  Исходя из графиков, мы решили состредоточиться на следующих кластерах:
  
  * Кластер `3`
    Его клиенты - это преимущественно участники промо-акции "Приведи друга". Они молоды (возраст от 20 лет), работают или живут неподалёку, оформляют абонементы на месяц и предпочитают не сильно тратиться на дополнительные услуги. Их "время жизни" - 1-3 месяца. Этот же кластер является и самым многочисленным относительно других.
  
  * Кластер `0`
    Почти все эти клиенты - участники партнёрской программы. Они взрослые (30+ лет), оформляют абонементы именно на год и дольше всех сотрудничают с фитнес-клубом. Эта группа - вторая по численности.
  
  * Кластер `4`
    Его клиенты по профилю похожи на клиентов кластера выше, но с некоторами исключениями: они старше (35+ лет), оформляют абонементы на полгода и больше прочих тратят на дополнительные услуги. Их численность лишь немногим меньше численности кластера `0`
  
  Кластеры `1` и `2` оказались промежуточными между этими тремя. Отличает их лишь отсутствие контактного телефона (в первом случае) и проживание/работа вдали от клуба (во втором).
  
  
* *Доли оттока по кластерам*

  **Кластеры `0` и `4`** по праву можно назвать **наиболее лояльными** клиентами - у них менее 10% оттока. А **у кластера `3` ситуация противоположная** - более 50% клиентов перестали посещать клуб в прошедшем месяце, и никаких гарантий на их возвращение (без каких-либо усилий со стороны клуба) нет.
  
Исходя из вышеизложенного разумно выдвинуть следующие **рекомендации**:

* Клиентам из кластера `4` имеет смысл предложить дополнительные скидки на годовые абонементы. Ведь чем дольше удерживается клиент, тем более он лоялен. Данная группа отличается высокой платёжеспособностью, но при этом чаще оформляет абонементы лишь на полгода. В годовой абонемент возможно включить и спец.предложения в виде дополнительных услуг (массажный салон, косметические процедуры), поскольку данные клиенты наверняка в них заинтересованы.

* Для клиентов из кластера `0` имеет смысл предложить пробные дополнительные услуги. Или, как вариант, расширить их годовые абонементы аналогичными спец.предложениями. Это мотивирует их посещать фитнес-клуб не дважды в неделю, а трижды (как это делают клиенты из предыдущего кластера).

* С кластером `3` сложнее всего. Он самый многочисленный - и одновременно с этим наименее лояльный. По результатам проведённого исследования можно предположить, что акция "Приведи друга" была конечной и длилась около трёх месяцев. Возможно, имеет смысл обсудить возможность её продления и увеличения её охвата. Если, к примеру, она действовала только для участников партнёрской программы, возможно стоит расширить её условия на всех действующих клиентов клуба. Помимо этого, потенциальных клиентов могут привлечь и бесплатные пробные занятия, проходящие в группах. Мы бы посоветовали не смешивать пробные группы с уже имеющимися, а проводить отдельные сессии - ввиду того, что возрастной диапазон в кластерах всё же различается.
