# Исследование популярности игр (Сборный проект)

## Описание проекта
Мы работаем в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Цель исследования:**
*Выявить критерии наиболее популярных игр.*

**Описание данных**

* *Name* — название игры
* *Platform* — платформа
* *Year_of_Release* — год выпуска
* *Genre* — жанр игры
* *NA_sales* — продажи в Северной Америке (миллионы проданных копий)
* *EU_sales* — продажи в Европе (миллионы проданных копий)
* *JP_sales* — продажи в Японии (миллионы проданных копий)
* *Other_sales* — продажи в других странах (миллионы проданных копий)
* *Critic_Score* — оценка критиков (максимум 100)
* *User_Score* — оценка пользователей (максимум 10)
* *Rating* — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

*Примечание: данные за 2016 год могут быть неполными.*

**Ход исследования**

Предварительно разделим исследование на четыре этапа:
 1. Обзор данных
 2. Предобработка данных и проведение расчётов
 4. Исследовательский анализ данных
 5. Составление портрета пользователя по регионам
 5. Проверка гипотез
 
О качестве данных ничего не известно. Поэтому до выполнения расчётов, проведения исследовательского анализа и проверки гипотез понадобится обзор данных. Мы проверим данные на ошибки и оценим их влияние на исследование. Затем, на этапе предобработки мы поищем возможность исправить самые критичные ошибки данных.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import math
from scipy import stats as st
pd.set_option('display.max_columns', None) # выводим все столбцы в таблице
```

Прочитаем файл `games.csv` из папки `/datasets`, сохраним его в переменной `df`и отобразим первые пять его строк:

```python
df = pd.read_csv('/datasets/games.csv')
display(df.head())
```

Для получения более полной информации о датафрейме вызовем метод `info()`:

```python
df.info()
```

Данные об играх представлены как категориальными (платформа, год выхода, жанр), так и количественными величинами (рейтинги и продажи). Всего в датафрейме одиннадцать столбцов, которые не поименованы в соответствии с общепринятым стилем. Представленные типы данных - `float` и `object`.

Количество значений в столбцах различается. Значит, в данных есть пропуски, которые необходимо исследовать.

### Промежуточный вывод
В каждой строке таблицы — информация об играх разных лет, взятая из открытых источников. Всего представлено одиннадцать столбцов.

Предварительно можно утверждать, что, данных достаточно для дальнейшего анализа параметров, определяющих игровую популярность. Однако, во многих столбцах встречаются пропуски. До этапов расчётов и исследовательского анализа необходимо заполнить пропущенные значения - или обосновать их отсутствие.

Для облегчения работы с данными необходимо привести все наименования столбцов к общепринятому стилю и провести проверку на дубликаты.

Интерес представляют и типы данных:
* в столбце `Year_of_Release` указан вещественный тип данных - его оптимальнее перевести в целочисленный;
* столбец `User_Score` лучше перевести в вещественный формат, так будет удобнее производить подсчёты и сравнения со столбцом `Critic_Score`.

Итак, чтобы двигаться дальше, нужно устранить проблемы в данных.

## Предобработка данных

### Стиль заголовков
Поскольку все столбцы имеют общую проблему в наименованиях (написаны заглавными буквами с нижними подчёркиваниями), их удобнее всего перевести к общепринятому стилю при помощи строчного метода `lower()`:

```python
df.columns = df.columns.str.lower()
```

Проверим результат при помощи `columns`:

```python
df.columns
```

### Изменение типов данных
Для начала переведём значения столбца `year_of_release` в целочисленный формат:

```python
df['year_of_release'] = df.year_of_release.astype('Int64')
```

Выведем тип данных столбца на экран:

```python
df.year_of_release.dtypes
```

Теперь рассмотрим столбец `user_score` на предмет того, какие значения он содержит:

```python
df.user_score.value_counts()
```

Выше фигурирует 2424 значения `tbd`. Исходя из терминологии игровой индустрии, аббревиатура переводится как "to be determined" и используется для обозначения того, что некий параметр (в данном случае, пользовательский рейтинг) еще не определён и будет объявлен позднее. Иными словами, для нашего анализа это значение эквивалентно пропуску.

Заменим `tbd` в столбце на `np.NaN`:

```python
df['user_score'] = df.user_score.replace('tbd', np.NaN)
```

Изменим тип данных в столбце на вещественный и выведем его на экран:

```python
df['user_score'] = df.user_score.astype('float')
df.user_score.dtypes
```

На этом изменение типов данных можно завершить и перейти к следующему шагу.

### Обработка пропусков
Выше мы уже обращали внимание на содержание пропущенных значений в столбцах. Выведем общее количество пропусков по датафрейму на экран:

```python
df.isna().sum()
```

Поочерёдно изучим каждый из столбцов и примем решение относительно пропусков - оставлять ли их либо предложить вариант замены.

Начнём рассмотрение со столбцов `name` и `genre`. Они содержат всего по два пропуска в каждом, носящие, по всей видимости, сугубо технический характер. Поэтому для удобства восприятия мы предпочтём заменить их на пустые строки `' '`:

```python
df['name'] = df.name.fillna(' ')
df['genre'] = df.genre.fillna(' ')
```

Проверим заполнение пропусков:

```python
df.isna().sum()
```

Далее оценим процент пропущенных значений в столбцах:

```python
pd.DataFrame(round(df.isna().mean()*100,)).style.background_gradient('coolwarm')
```

Отсутствие года выхода `year_of_release` можно обосновать технологическими причинами. В теории мы бы могли предложить вариант замены для этого столбца, исходя из предположения о том, что одна и та же игра могла выходить на разных платформах, - и заполнить пропуски в годе выхода игры на одной платформе известным годом выхода той же игры на другой. Но здесь есть несколько ограничений:
* игра могла выходить как эксклюзив для одной платформы
* процесс кроссплатформенной эмуляции мог идти длительное время (до года включительно)

Но главной причиной против заполнения пропусков в этом столбце будет всё же факт их малого присутствия в текущей выборке (2%). Поэтому мы предпочтём оставить их без изменения.

Выведем на экран строки, содержащие пропуски в `critic_score`,`user_score` и `rating`, и взглянем, есть ли пересечения между ними:

```python
display(df.query('critic_score.isna() & rating.isna() & user_score.isna()'))
```

Мы видим 6667 строк с отсутствующими значениями оценок пользователей, критиков и рейтинга ESRB. Причём, год выхода игры значения не имеет - присутствуют как неоценённые игры 1980х годов, так и недавние, 2016 года выпуска.

Причины появления пропусков могут быть самыми разными, но вряд ли они являются полностью технологическими. По идее, можно попытаться выдвинуть так называемую "теорию ранних игр": *игры до 1995 года (условно) могут не иметь рейтингов вовсе*. Развитие этой теории приведено ниже.

Если мы посмотрим, за какие годы представлены игры в нашем датафрейме, то увидим, что самые ранние датируются 1980 годом. Основные игровые обозреватели (Игромания, IGN и Metacritic) появились значительно позже - только во второй половине девяностых. Легко предположить, что если рейтинги игр и составлялись ранее, то они носили сугубо локальный характер, и вряд ли представлены в текущем датафрейме. Касательно пользовательских отзывов можно отметить то, что до того же времени интернет был гораздо менее распространён в принципе, поэтому обмен мнениями касательно игр (а уж тем более - сетевые опросы на данную тему) за те годы ничтожно мал. Организация ESRB, чей рейтинг указан в столбце `rating`, была основана аж в 1994 году, поэтому данные о вышедших ранее играх справедливо отсутствуют. К тому же, ESRB не является единственной организацией, присуждающей рейтинг - есть ещё PEGI, CERO и многие другие. 

Также в описании проекта говорится, что данные за 2016 год представлены не полностью, что также намекает на обоснованное наличие пропущенных значений.

Как бы то ни было, в столбцах с рейтингами и оценками пропуски составляют практически половину от всех содержащихся там данных. Сбрасывать их было бы грубейшей ошибкой - равно как и заполнять чем-либо осмысленным. Ни среднее, ни медианное значение (пусть даже и взятое по какой-либо группе) неправдоподобно исказит наши данные и дальнейшие расчёты по ним. Всё-таки, формирование любого рода рейтингов слишком индивидуально, и в текущем исследовании мы только будем смотреть на параметры, предположительно их определяющие.

Тем не менее, из этой патовой ситуации есть выход - заменить пропуски на значения-заглушки. Так одновременно и поймём, сколько пропусков содержится в столбцах, и не исказим итоговую картину, если в дальнейшем поставим фильтр на эти значения. Осуществлять замену мы будем только в столбце `rating` - пропуски `critic_score` и `user_score` для сохранения правдоподобности оставим без изменения.


Заменим пропущенные значения в столбце `rating` на `unknown` и взглянем на количество пропусков после внесённых правок:

```python
df['rating'] = df.rating.fillna('unknown')
df['critic_score'] = df.critic_score.fillna(-1)
df['user_score'] = df.user_score.fillna(-1)

df.isna().sum()
```

Результат нас устраивает, можно переходить к следующему этапу.

### Проверка на дубликаты
Посмотрим, присутствуют ли в датафрейме явные дубликаты:

```python
df.duplicated().sum()
```

Дублированных строк не обнаружено.

Далее проверим на наличие неявных дубликатов столбцы с категориальными значениями. Начнём с `genre` - и выведем все его уникальные значения:

```python
df.genre.unique()
```

Все значения уникальны, повторов не обнаружено.

Аналогичным образом проверим столбец `platform`:

```python
df.platform.unique()
```

В категориях выше повторы также отсутствуют.

И наконец, проверим столбец `rating`, поскольку он также содержит категориальные значения:

```python
df.rating.unique()
```

Выше представлены все рейтинги игр ESRB. Согласно данным сайта ESRB (https://www.esrb.org/history/), рейтинг "Е" ("Everyone") до 1998 года носил название "K-A" ("Kids to Adults"). По этой причине мы можем осуществить замену:

```python
df['rating'] = df.rating.replace('K-A', 'E')
df.rating.unique()
```

На этом процесс предобработки данных можно считать завершённым. Но прежде, чем перейти к исследовательскому анализу данных, требуется выполнить дополнительный расчёт ниже.

### Расчёт суммарных продаж
Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец `total_sales` и выведем на экран пять первых строк обновлённого датафрейма:

```python
df['total_sales'] = df[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
display(df.head())
```

### Промежуточный вывод
Предобработка выявила следующие проблемы в данных:

* некорректный стиль представления заголовков

* неудобный тип представления данных
* дубликаты

Были внесены изменения:

* все заголовки были приведены к общепринятому стилю ("snake_case");


* тип данных в столбце `year_of_release` изменён на целочисленный (`Int64`); тип данных в `user_score` заменён на вещественный;


* значения `tbd` в столбце `user_score` заменены на `np.NaN`;


* пропущенные значения в следующих столбцах заменены на значения-заглушки:
   * в `name` - на пустую строку `' '`
   * в `genre` - на пустую строку `' '`
   * в `rating` - на `unknown` 
   * в `user_score` - на `-1`
   * в `critic_score` - на `-1`


* пропущенные значения `year_of_release` были оценены и, с целью сохранения правдоподобности данных, оставлены без изменения;


* выявлены и заменены неявные дубликаты в столбце `rating` (`K-A` изменено на `E`);


* посчитаны суммарные продажи во всех регионах, данные внесены в столбец `total_sales`.
  
В процессе предобработки нам не встретились явные аномалии (к примеру, отрицательные значения) в данных. Вероятно, они отсутствуют.

### Влияние отзывов на продажи
Сперва посмотрим, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Изучим это на примере одного из лидеров - PlayStation 4.

Выберем срез данных для этой консоли и построим попарные диаграммы рассеяния для оценок и продаж:

```python
print('PlayStation 4 scatter matrix')
sns.pairplot(actual_platforms.query(
    'platform == "PS4" and critic_score >= 0 and user_score >= 0'
)[['critic_score', 'user_score', 'total_sales']])
plt.show()
```

Выше мы наблюдаем, что:
* есть предположительная взаимосвязь между отзывами критиков и пользователей (мнение первых влияет на формирование мнение вторых, и наоборот);
* явной линейной связи между отзывами и продажами не прослеживается, однако есть случаи высоких продаж для игр с высокой оценкой;
* судя по распределению продаж, в выборке преобладают игры с невысокими продажами.

Для уточнения выведем матрицу корреляций по рассматриваемым параметрам:

```python
print("Correlation coefficients")
actual_platforms.query('platform == "PS4" and critic_score >= 0 and user_score >= 0')[['critic_score', 'user_score', 'total_sales']].corr()
```

Можно заключить следующее:
* линейная взаимосвязь между оценками критиков и оценками пользователей определённо присутствует;
* на продажи в некотором роде влияют отзывы критиков, но они не являются полностью определяющим фактором;
* взаимосвязь продаж с отзывами игроков либо отсутствует, либо не носит линейный характер.

Соотнесём выводы выше с продажами игр на других платформах. Для начала выведем срезы данных по интересующим нас платформам, применив фильтр на значения-заглушки:

```python
ps4 = actual_platforms.query('user_score != -1 & platform == "PS4" & critic_score != -1')
x_one = actual_platforms.query('user_score != -1 & platform == "XOne" & critic_score != -1')
nin_3ds = actual_platforms.query('user_score != -1 & platform == "3DS" & critic_score != -1')
nin_wiiu = actual_platforms.query('user_score != -1 & platform == "WiiU" & critic_score != -1')
```

Построим графики по платформам Xbox One, Nintendo WiiU и Nintendo 3DS и определим степень линейной корреляции параметров:

```python
# вычисляем коэффициенты корреляции
print('Корреляция Xbox One пользователи:', x_one['user_score'].corr(x_one['total_sales']).round(2))
print('Корреляция Nintendo 3DS пользователи:',nin_3ds['user_score'].corr(nin_3ds['total_sales']).round(2))
print('Корреляция Nintendo WiiU пользователи:',nin_wiiu['user_score'].corr(nin_wiiu['total_sales']).round(2))

# строим графики зависимостей от отзывов пользователей
ax = ps4.plot(x = 'user_score', y='total_sales', style='o', label='PS4', alpha=0.9)
x_one.plot(x = 'user_score', y='total_sales', style='x', ax=ax, label='XboxOne', alpha=0.7)
nin_3ds.plot(kind = 'scatter', x = 'user_score', y='total_sales', color = 'green', ax=ax,  label='3DS', alpha=0.7)
nin_wiiu.plot(kind = 'scatter', x = 'user_score', y='total_sales', color = 'red', ax=ax,  label='WiiU', alpha=0.7)
plt.title('График зависимости продаж платформ от оценки пользователей', size=10);
plt.xlabel('Оценка пользователей')
plt.ylabel('Кол-во продаж (млн копий)')
plt.show()
```

```python
# вычисляем коэффициенты корреляции
print('Корреляция Xbox One критики:', x_one['critic_score'].corr(x_one['total_sales']).round(2))
print('Корреляция Nintendo 3DS критики:',nin_3ds['critic_score'].corr(nin_3ds['total_sales']).round(2))
print('Корреляция Nintendo WiiU критики:',nin_wiiu['critic_score'].corr(nin_wiiu['total_sales']).round(2))

# строим графики зависимостей от отзывов критиков
ax = ps4.plot(x = 'critic_score', y='total_sales', style='o', label='PS4', alpha=0.9)
x_one.plot(x = 'critic_score', y='total_sales', style='x', ax=ax, label='XboxOne', alpha=0.7)
nin_3ds.plot(kind = 'scatter', x = 'critic_score', y='total_sales', color = 'green', ax=ax,  label='3DS', alpha=0.7)
nin_wiiu.plot(kind = 'scatter', x = 'critic_score', y='total_sales', color = 'red', ax=ax,  label='WiiU', alpha=0.7)
plt.title('График зависимости продаж платформ от оценки критиков', size=10);
plt.xlabel('Оценка критиков')
plt.ylabel('Кол-во продаж (млн копий)')
plt.show()
```

Отметим некоторые особенности:

**Xbox One**
* общая картина идентична ситуации с PlayStation 4, однако в этом случае отзывы критиков оказывают чуть более весомое влияние на продажи;


**Nintendo WiiU**
* здесь обе зависимости можно охарактеризовать как умеренные - т.е. отзывы всё же оказывали какое-то влияние на продажи игр для этой платформы;

**Nintendo 3DS**
* в данном случае демонстрируется слабая линейная зависимость как от отзывов пользователей, так и от отзывов критиков - по всей вероятности, продажи были обусловлены иными критериями.

Во всех перечисленных случаях стоит сделать поправку на неполные сведения за 2016 г. и на развитие платформ нового поколения в то время.

## Проверка гипотез
Перед нами стоит задача - проверить предположения, звучащие как:
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Мы можем воспользоваться статистическим методом и сформулировать нулевые и альтернативные гипотезы для обоих рассматриваемых случаев на основе статистической гипотезы о равенстве средних двух генеральных совокупностей. Сравнение будет производиться при помощи t-теста.

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
Прежде, чем формулировать нулевую и альтернативную гипотезы для данного случая, требуется изучить аспекты, необходимые для проведения t-теста:

* *Генеральные совокупности не должны зависеть друг от друга.*
  
  В данном случае пользователи выбирали платформы, исходя из собственных предпочтений, доступности и финансовых возможностей, поэтому рассматриваемые генеральные совокупности друг от друга не зависят.
  
  
* *Выборочные средние должны быть нормально распределены.*

   Поскольку рассматриваемые данные довольно обширны, а величины являются непрерывными, то можно предположить, что в подобных ситуациях распределение всегда будет стремиться к нормальному.
   
   
* *Дисперсии рассматриваемых генеральных совокупностей должны быть равны.*

   В этом случае мы не можем утверждать наверняка, необходимо провести проверку. Для начала выделим два среза данных по рейтингам пользователей - `sample_1` и `sample_2` - для платформ `XOne` и `PC`, соответственно, и сбросим в них все пропуски, которые помешают нам в расчётах:

```python
sample_1 = actual_platforms.query('platform == "XOne" & user_score != -1').user_score
sample_2 = actual_platforms.query('platform == "PC" & user_score != -1').user_score
```

Оценим дисперсии "на глаз" путём простого сравнения:

```python
print(f'Дисперсия по XOne: {sample_1.describe()["std"]**2}\nДисперсия по PC: {sample_2.describe()["std"] **2}')
```

Дополнительно проведём тест Левене на предмет оценки равенства дисперсий:

```python
stat, p = st.levene(sample_1, sample_2)
p
```

Значение вероятности выше достаточно велико. Т.е. хоть дисперсии и не идентично равны друг другу, но из значения всё же достаточно близки, чтобы считать их равными. В данном случае мы примем для t-теста параметр `equal_var = True` (значение по умолчанию).

Таким образом, нулевая гипотеза будет звучать так:

*Средние пользовательские рейтинги платформ Xbox One и PC одинаковые*

Альтернативная гипотеза будет звучать следующим образом:

*Средние пользовательские рейтинги платформ Xbox One и PC отличны друг от друга*

Критический уровень значимости `alpha` примем равным 0.05, поскольку высокая точность в конкретном исследовании нам не требуется. Если подсчитанная вероятность окажется ниже этого порогового значения, то это будет достаточным основанием, чтобы отвергнуть нулевую гипотезу:

```python
alpha = .05

results = st.ttest_ind(sample_1, sample_2)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Это можно интерпретировать как "вероятность получить наблюдаемый результат при условии того, что нулевая гипотеза верна, равна 28%". Т.е. выборочный средний рейтинг пользователей за период 2015-2016 гг. по платформе Xbox One с вероятностью 28% равен такому же рейтингу для ПК за аналогичный период.

### Средние пользовательские рейтинги жанров Action и Sports разные
Аналогично изучим ниже аспекты для проведения t-теста:

* *Выборочные средние должны быть нормально распределены.*

* *Дисперсии рассматриваемых генеральных совокупностей должны быть равны.*

   В этом случае мы также не можем сказать наверняка, поэтому проведём промежуточную проверку. Для начала выделим два среза данных по рейтингам пользователей - `sample_3` и `sample_4` - для жанров `Action` и `Sports`, соответственно, и сбросим в них все пропуски, которые помешают нам в расчётах:

Оценим дисперсии "на глаз" путём простого сравнения:

Значение вероятности выше очень и очень мало, а значит, и дисперсии не могут считаться равными. В данном случае мы примем для t-теста параметр `equal_var = False`.

Таким образом, нулевая гипотеза будет звучать так:

*Средние пользовательские рейтинги жанров Action и Sports одинаковые*

Альтернативная гипотеза будет звучать следующим образом:

*Средние пользовательские рейтинги жанров Action и Sports отличны друг от друга*

Критический уровень значимости `alpha` примем равным 0.05, поскольку высокая точность в конкретном исследовании нам не требуется. Если подсчитанная вероятность окажется ниже этого порогового значения, то это будет достаточным основанием, чтобы отвергнуть нулевую гипотезу:

```python
alpha = .05

results = st.ttest_ind(sample_3, sample_4, equal_var = False)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Это можно интерпретировать как "вероятность получить наблюдаемый результат при условии того, что нулевая гипотеза верна, равна 0.0000000597%". Т.е. выборочный средний рейтинг пользователей жанра "Action" за период 2015-2016 гг. отличен от рейтинга по жанру "Sports" за аналогичный период. Но, поскольку рассматриваемая гипотеза была двусторонней, то мы не можем точно сказать, в какую сторону расположено это отличие.
